# Instalación local de un RDBMS (Windows)

Hay dos maneras de acceder a manejadores de bases de datos:

* Instalar en máquina local un administrador de bases relacional.
* Tener ambientes de desarrollo especiales o servicios cloud.

Usaremos MySQL porque tiene un impacto histórico siendo muy utilizado y además es software libre y gratuito. 

**Root** es el usuario principal que tendrá todos los permisos y por lo tanto en ambientes de producción hay que tener mucho cuidado al configurarlo.

### 1. Enlace al instalador de MySQL versión 8.0.31

https://dev.mysql.com/downloads/installer/

Elegimos la opción dos (subrayada), ya que es el instalador comppleto.




### 2. Descargar el instalador

No necesitamos iniciar sesión, y solo damos click en la opción que dice "no gracias, solo quiero la descarga"



### 3. Ejecutar el instalador

* Una vez se haya completado la descarga del instalador lo ejecutamos como administrador



* Tras darle los permisos necesarios de instalación nos aparecerá la siguiente ventana con los diferentes tipos de instalación, en nuestro caso elegimos la opción **Custom** porque existe componentes que no vamos a utilizar.



* Al darle en la botón siguiente nos aparece la ventana con todos los componentes que podemos instalar, le damos click en el símbolo más de la opción **MySQL Server** y abrimos las pestañas hasta que nos aparezca de la siguiente manera.



* Con la flecha pasamos la opción que indica el instalador para 64 bits.



* Seguimos el paso anterior con la opción **Applications** y posterior en **MySQL Workbentch 8.0** que es un cliente visual que ayuda a realizar las tareas de Bases de Datos de forma mucho más sencilla y gráfica. Finalmente lo pasamos con la flecha para su instalación.



* Finalmente me muestra una ventana que indica que los componentes seleccionados están listos para instalar, así que unicamente damos click en **Ejecutar**.



* Una vez que haya finalizado la instalación nos aparacerá la siguiente ventana si no ocurrió ningún problema.



* Le damos en **Siguiente** y nos indica que el servidor esta listo para instalar. 



* Nuevamente damos click en **Siguiente** y nos aparece el tipo de configuración que buscamos, escogemos la opción por default de Computadora para Desarrollo. Además nos aparece el puerto en el que se va a instalar, que más adelante nos va a pedir para instalar el componente visual, sin más damos click en **Siguiente**. 



* En la siguiente ventana mantenemos el método de autenticación por defecto y damos en **Siguiente**.



* A continuación, en la ventana de Cuentas y Roles y en **Root Password** que es el usuario principal de la base de datos que tiene todos los permisos, asigna una contraseña segura y que recuerdes porque la vas a utilizar cada que te conectes con este usuario. 



* Al dar click en siguiente nos pregunta si queremos confirgurarlo como servicio, para que corra en el Background de Windwos para que no lo estemos corriendo cada vez y también si se quiere que inicie cada vez que se incie el computador. 



* Ya para terminar nos muestra un check list que va a ir recorriendo conforme se vaya ejecutando el proceso de isntalación.



* Damos click en **Terminar** y nos indica que el producto a sido configurado.



* Damos click en **Siguiente** y nos dice que toda la instalación del servidor como del cliente visual esta correcta. Finalmente damos click en **Terminar**
